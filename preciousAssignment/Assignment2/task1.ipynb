```python
def mean(x):
    i=0
    for n in x:
        #adding the next number to the initial
        i+=n
    #mean of x(meanx) is the new initial(all the addes numbers) divided by the length of the list(x)
    mean=i/len(x)
    return mean
```

```python
#FR2: Develop a function to read a single specified column of data from a CSV file
def read_file(file, column):
    #opening file
    file = open(file)
    li=[]
    #Looping through the file columns and replacing \n with nothing and\t with commas and then splits the list by the commas
    for i in file:
        i=i.replace("\n", "").replace("\t", ",").split(",")
        li.append(i)

    column_val=[]
    #looping through the list
    for i in range(len(li)):
        if i == 0:
            colmn_key=li[i][column]
        elif i>0:
            column_val.append(float(li[i][column]))
    all={colmn_key:column_val}
    return all
```

```python
#FR3 Develop a function to read CSV data from a file into memory

def data(file):
    e=open(file)
    li = []
    for i in e:
        i=i.replace("\n", "").replace("\t", ",").split(",")
        li.append(i)
    ncol=len(li[0])

    for i in range(ncol):
        dict_data={}
        for col in range(ncol):
            all = read_file(file, col)
            dict_data.update(all)
        return dict_data
```

```python
#FR4 Develop a function to calculate the Pearson Correlation Coefficient for two lists of data 
def pearson_coefficient(list1, list2):
    def mean_diff(x):
        list=[]
        for i in x:
            new_diff=i-mean(x)
            list.append(new_diff)
        return list

    def sum(list):
        initial_num=0
        for i in list:
            initial_num=i+initial_num
        return initial_num
    
    def mul(list1, list2):
        new_list=[]
        inc_fac=0

        for i in range(len(list1)):
            for j in range(len(list2)):
                if (i==inc_fac) and (j==inc_fac):
                    x=list1[inc_fac]*list2[inc_fac]
                    new_list.append(x)
                    inc_fac+=1
        return new_list

    def power(x,y):
        return x**y
    
    def sqrt(x):
        return x**0.5

    pearson_cor_coeff = sum(mul(mean_diff(list1), mean_diff(list2)))/sqrt(sum(mul(mean_diff(list1), mean_diff(list1)))*sum(mul(mean_diff(list2), mean_diff(list2))))
    return pearson_cor_coeff
```

```python
#FR5 Develop a function to generate a set of Pearson Correlation Coefficients for a given data file
def readpearson(file):
    new_data=[]
    for i in data(file):
        for j in data(file):
            if i==j:
                new_tuple = (i, j, "----")
                new_data.append(new_tuple)
            else:
                new_tuple=(i, j, round(pearson_coefficient(data(file)[i], data(file)[j]), 4))
                new_data.append(new_tuple)

    return new_data
```
